```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np; np.random.seed(1)
from nltk.corpus import wordnet as wn
import os
from sklearn.manifold import TSNE
import sys
import glob

lemma_prefix = 'LEMMA:'
synset_prefix = 'SYNSET:'
instance_prefix = 'INSTANCE:'

def populate_vocab_dict():
    vocab_file = '/usr/local/google/home/agoldie/data/uncased_L-12_H-768_A-12/vocab.txt'
    vocab_dict = {}
    with open(vocab_file, 'r', errors='replace') as f:
        for i, word in enumerate(f.readlines()):
            vocab_dict[i] = word.strip()
    return vocab_dict

def get_synsets(word, pretty_print=False):
    return [synset.name() for synset in wn.synsets(word)]

def parse_file(file_name, word, vocab_dict, max_words=sys.maxsize):
    with open(file_name, 'r') as f:
        positions = []
        contexts = []
        embeddings = []
        #for line in f.readlines():
        for i, line in enumerate(f):
            if i == max_words:
                break
            #pos, context, embedding_line = line.split('\t')
            segments = line.split('\t')
            if len(segments) == 3:
                pos, context, embedding_line = segments
            else:
                pos, context, embedding_line, shard = segments
            pos = int(pos)
            context = [int(i) for i in context.split(' ')]
            context = [vocab_dict[word_id] for word_id in context]
            new_context = []
            shift = 1
            for i in range(1, len(context)):
                prev_word = context[i-1]
                curr_word = context[i]
                if curr_word.startswith('##'):
                    if i < pos:
                        shift += 1
                    partial_word = new_context[-1] if new_context else ''
                    new_context[-1] = partial_word + curr_word.strip('#')
                else:
                    new_context.append(curr_word)
            context = new_context
            #context = " ".join(human_readable)
            contexts.append(context)
            embedding = [float(i) for i in embedding_line.split(' ')]
            embeddings.append(embedding)
            pos -= shift
            positions.append(pos)
    return positions, contexts, embeddings

def update_annot(ind):
    pos = sc.get_offsets()[ind["ind"][0]]
    annot.xy = pos
    text = " ".join(list(map(str,ind["ind"])))
    # TODO: Update this to handle lists.
    example_index = ind["ind"][0]
    
    pos, context = pos_and_contexts[example_index]
    before_text = ' '.join(context[:pos])
    before_text = before_text.strip('[CLS]').strip()
    after_text = ' '.join(context[pos+1:])
    after_text = after_text.split('[SEP]')[0]
    text = before_text + r" $\bf{" + context[pos] + "}$ " + after_text
    
    annot.set_text(text)
    annot.get_bbox_patch().set_alpha(0.4)


def hover(event):
    vis = annot.get_visible()
    if event.inaxes == ax:
        cont, ind = sc.contains(event)
        if cont:
            update_annot(ind)
            annot.set_visible(True)
            fig.canvas.draw_idle()
        else:
            if vis:
                annot.set_visible(False)
                fig.canvas.draw_idle()

def get_embedding_file(word):
    if word != '.' and '.' in word:
        data_dir = '/usr/local/google/home/agoldie/data/wordnet/words/'
        data_dir = '/tmp/test/words/'
#     if word != '-' and '-' in word and '_' in word:
#         data_dir = '/usr/local/google/home/agoldie/data/gigaword/sentences/embeddings/'
    else:
        data_dir = '/usr/local/google/home/agoldie/data/gigaword/words/'
    file_name = os.path.join(data_dir, word + '.txt')
    return file_name
 
def process_file(file_name, vocab_dict, max_words=250):
    with open(file_name, 'r') as f:
        positions = []
        contexts = []
        embeddings = []
        #for line in f.readlines():
        for i, line in enumerate(f):
            if i == max_words:
                break
            #pos, context, embedding_line = line.split('\t')
            segments = line.split('\t')
            if len(segments) == 3:
                pos, context, embedding_line = segments
            else:
                pos, context, embedding_line, shard = segments
            pos = int(pos)
            context = [int(i) for i in context.split(' ')]
            context = [vocab_dict[word_id] for word_id in context]
            new_context = []
            shift = 1
            for i in range(1, len(context)):
                prev_word = context[i-1]
                curr_word = context[i]
                if curr_word.startswith('##'):
                    if i < pos:
                        shift += 1
                    partial_word = new_context[-1] if new_context else ''
                    new_context[-1] = partial_word + curr_word.strip('#')
                else:
                    new_context.append(curr_word)
            context = new_context
            #context = " ".join(human_readable)
            contexts.append(context)
            embedding = [float(i) for i in embedding_line.split(' ')]
            embeddings.append(embedding)
            pos -= shift
            positions.append(pos)
    return positions, contexts, embeddings

def generate_wordnet_embedding(word, vocab_dict):
    wordnet_dir = '/usr/local/google/home/agoldie/data/wordnet/words/'
    pattern = wordnet_dir + '*#{}.txt'.format(word)
    files = glob.glob(pattern)
    embeddings = []
    for file in files:
        _, _, e = process_file(file, vocab_dict)
        embeddings.extend(e)
    positions = [0]
    synset = wn.synset(word)
    context = [synset.name().replace('_', '\_'), ':']
    definition = synset.definition()
    context.extend(definition.split(' '))
    contexts = [context]
    embeddings = np.array(embeddings).tolist()
    embeddings = [np.average(embeddings, axis = 0)]
    return positions, contexts, embeddings
    
# TODO: Handle missing words, handle '.'
def parse_embeddings(word, max_words=sys.maxsize):
    word_type = ''
    vocab_dict = populate_vocab_dict()
    file_name = get_embedding_file(word)
    if word.startswith(synset_prefix):
        word_type = synset_prefix[:-1]
        word = word.split(synset_prefix)[1]
        subdir = 'wordnet'
        return generate_wordnet_embedding(word, vocab_dict)
    elif word.startswith(lemma_prefix):
        word_type = lemma_prefix[:-1]
        word = word.split(lemma_prefix)[1]
        subdir = 'gigaword'
        #gigaword_dir = '/usr/local/google/home/agoldie/data/gigaword/words/'
        #file_name = os.path.join(gigaword_dir, word + '.txt')
        #file_name = '/tmp/test_tree_output/words/' + word + '.txt'
    elif word.startswith(instance_prefix):
        word_type = instance_prefix[:-1]
        word = word.split(instance_prefix)[1]
        subdir = 'wordnet'
    else:
        subdir = 'gigaword'
    file_name = '/usr/local/google/home/agoldie/data/{}/words/{}.txt'.format(subdir, word)
    positions, contexts, embeddings = process_file(file_name, vocab_dict)
    if word_type == instance_prefix[:-1]:
        synset_name = word.split('#')[1]
        for i in range(len(positions)):
            positions[i] += 2
            contexts[i].insert(0, ':')
            contexts[i].insert(0, synset_name)
    return positions, contexts, embeddings

    
def extract_wordnet_senses(word):
    senses = []
    wordnet_dir = '/usr/local/google/home/agoldie/data/wordnet/words/'
    for synset in wn.synsets(word):
        name = synset.name()
        for lemma in synset.lemmas():
            word_sense = lemma.name() + '#' + name
            file = os.path.join(wordnet_dir, word_sense + '.txt')
            if os.path.exists(file):
                senses.append(word_sense)
    return senses

def extract_lemmas(word):
    lemmas = set()
    gigaword_dir = '/usr/local/google/home/agoldie/data/gigaword/words/'
    for synset in wn.synsets(word):
        name = synset.name()
        for lemma in synset.lemmas():
            lemma = lemma.name()
            file = os.path.join(gigaword_dir, lemma + '.txt')
            if os.path.exists(file):
                lemmas.add('LEMMA:' + lemma)
    return lemmas

# This just doesn't seem correct, I should make sure about it. Why was it labeled as gigaword_dir?
# Why is this seemingly identical to extract_wordnet_senses()?
def extract_instances(word):
    instances = set()
    wordnet_dir = '/usr/local/google/home/agoldie/data/wordnet/words/'
    for synset in wn.synsets(word):
        synset_name = synset.name()
        for lemma in synset.lemmas():
            lemma = lemma.name()
            instance = lemma + '#' + synset_name
            file = os.path.join(wordnet_dir, instance + '.txt')
            if os.path.exists(file):
                instances.add('INSTANCE:' + instance)
    return instances

def extract_synsets(word):
    return ['SYNSET:' + s.name() for s in wn.synsets(word)]
```

```python
word = 'ugly'
lemmas = extract_lemmas(word)
num_lemmas = len(lemmas)

instances = extract_instances(word)
num_instances = len(instances)
words = []
words.extend(lemmas)
words.extend(instances)
        
tuples = [parse_embeddings(word, 250) for word in words]

positions = [t[0] for t in tuples]
contexts = [t[1] for t in tuples]
embeddings = [t[2] for t in tuples]

lengths = [len(e) for e in embeddings]
embeddings = np.concatenate(embeddings)

y = []
for i in range(len(words)):
    y.append(i * np.ones(lengths[i]))

pos_and_contexts = []
for i in range(len(words)):
    for (p, c) in zip(positions[i], contexts[i]):
        pos_and_contexts.append([p, c])

y = np.concatenate(y)  
X = np.array(embeddings) 

tsne = TSNE()
X_embedded = tsne.fit_transform(X)

norm = plt.Normalize(1,4)
cmap = plt.cm.Spectral
cmap = plt.cm.coolwarm
cmap = plt.cm.gist_rainbow

fig,ax = plt.subplots()
sc = plt.scatter(X_embedded[:,0], X_embedded[:,1], c=y, s=5, alpha=.75, cmap=cmap)

annot = ax.annotate("", fontsize=4, xy=(0,0), xytext=(-100,10),textcoords="offset points",
                    bbox=dict(boxstyle="round", fc="w"),
                    arrowprops=dict(arrowstyle="->"))

annot.set_visible(False)

fig.canvas.mpl_connect("motion_notify_event", hover)

legend_size = min(10, len(words))
handles, _ = sc.legend_elements(num=range(len(words)))
lemma_handles = handles[:num_lemmas]
instance_handles = handles[num_lemmas:]

labels = []
for word in words:
    if word.startswith(lemma_prefix):
        word = word[len(lemma_prefix):]
    if word.startswith(synset_prefix):
        word = word[len(synset_prefix):]
    if word.startswith(instance_prefix):
        word = word[len(instance_prefix):]
    labels.append(word)

lemma_labels = labels[:num_lemmas]
instance_labels = labels[num_lemmas:]

lemma_legend = ax.legend(lemma_handles, lemma_labels, loc='upper left', fontsize=4)
ax.add_artist(lemma_legend)
instance_legend = ax.legend(instance_handles, instance_labels, loc='lower right', fontsize=4)

plt.show()
```
